### below is a venn digram of datasets review in 4 english literature reviews of the Image Aesthetic Quality Assement Domain (IAQA)

```python
from venn import generate_petal_labels, draw_venn, generate_colors



dataset_dict = {
    'S. Kanwal et al.': hand_crafted,
    "H. Yang et al.": comprehensive_survey,
'Y. Deng et al.': experimental_survey,
'D. Spathis': photo_qualitiy_evaluation}
petal_labels = generate_petal_labels(dataset_dict.values())


#significance_filter = lambda value: value if int(value) >= 45 else "ns"
petal_labels = {
    logic: value
    for logic, value in petal_labels.items()
}

print(petal_labels)
petal_labels['0001']='\n'.join(photo_qualitiy_evaluation-hand_crafted-experimental_survey)

petal_labels['1111']='\n'.join(photo_qualitiy_evaluation&
                               hand_crafted&experimental_survey
                               &comprehensive_survey)
petal_labels['1001']='\n'.join(dataset_dict['D. Spathis']&dataset_dict['S. Kanwal et al.']-
                              dataset_dict['Y. Deng et al.']-dataset_dict['H. Yang et al.'])
petal_labels['0110']='\n'.join(dataset_dict['Y. Deng et al.']&dataset_dict['H. Yang et al.']
                              -(dataset_dict['D. Spathis']|dataset_dict['S. Kanwal et al.']))
petal_labels['1000']='\n'.join(dataset_dict['S. Kanwal et al.']-
                              (dataset_dict['D. Spathis']|dataset_dict['Y. Deng et al.']))
petal_labels['0100']='\n'.join(dataset_dict['H. Yang et al.']-
                              (dataset_dict['D. Spathis']|dataset_dict['Y. Deng et al.']))
petal_labels['1110']='\n'.join(dataset_dict['H. Yang et al.']&dataset_dict['S. Kanwal et al.']&
                               dataset_dict['Y. Deng et al.']-dataset_dict['D. Spathis'])
print(petal_labels)
fig.patch.set_facecolor((.88, .88, .88))

draw_venn(
    petal_labels=petal_labels, dataset_labels=dataset_dict.keys(),
    hint_hidden=False, colors=generate_colors(n_colors=4),
    figsize=(8, 8), fontsize=14, legend_loc="best", ax=None
)


plt.title("IAQA Datasets In Reviews", size=20, fontweight='bold',
         pad=20)

plt.savefig('datasets.png')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

from matplotlib_venn import venn3, venn3_circles

fig = plt.figure(figsize=(11,10))

fig.patch.set_facecolor((.88, .88, .88))
v = venn3([A,B,C], ('Computer Vision', 'Wider Aesthetics', 'Image Processing'),
         set_colors=( (1, 0, .2), (0, 1, 0.2) , (0, 0.1, 1)), 
 alpha = 0.5)
for text in v.set_labels:
   text.set_fontsize(25)

v.get_label_by_id('110').set_text('\n'.join(A&B-C))
v.get_label_by_id('011').set_text('\n'.join(B&C-A))
v.get_label_by_id('101').set_text('\n'.join(A&C-B))

v.get_label_by_id('100').set_text('\n'.join(A-B-C))
v.get_label_by_id('001').set_text('\n'.join(C-A-B))
v.get_label_by_id('010').set_text('\n'.join(B-A-C))

v.get_label_by_id('111').set_text('\n'.join(A&B&C))

c = venn3_circles(subsets=(A,B,C), linestyle='dashed')

plt.title("Inter-Relation Venn Digram of IAQA ", size=20, fontweight='bold',
         pad=20)
plt.annotate('Learning Aesthetic Attributes Only', xy=v.get_label_by_id('110').get_position() +
             np.array([0, 0.1]), xytext=(60,80), ha='center',
             textcoords='offset points', 
             bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.4),
             arrowprops=dict(arrowstyle='->',              
                             connectionstyle='arc',color='black'))
plt.annotate('Learning Using Coventional CV \n (non deep learning)', xy=v.get_label_by_id('101').get_position() +
             np.array([-0.12, -0.05]), xytext=(-140,-130), ha='center',
             textcoords='offset points', 
             bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.4),
             arrowprops=dict(arrowstyle='->',              
                             connectionstyle='arc',color='black'))
plt.annotate('Image Processing for Image Aesthetics \n (no Learning/Abstraction)', xy=v.get_label_by_id('011').get_position() +
             np.array([.1, -.1]), xytext=(120,-95), ha='center',
             textcoords='offset points', 
             bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.4),
             arrowprops=dict(arrowstyle='->',              
                             connectionstyle='arc',color='black'))

plt.show()
fname = 'inter-relation_of_subjects'
fig.savefig(fname, dpi=150,)
```

### For visualizing data type as Dedrogram and exploring hierarchies

```python
import networkx as nx
%matplotlib inline
G = nx.Graph()
G.add_edge(1, 2, color="red")
G.add_edge(2, 3, color="red")
G.add_node(3)
G.add_node(4)

A = nx.nx_agraph.to_agraph(G)  # convert to a graphviz graph
A.draw("attributes.png", prog="neato")  # Draw with pygraphviz

# convert back to networkx Graph with attributes on edges and
# default attributes as dictionary data
X = nx.nx_agraph.from_agraph(A)
print("edges")
print(list(X.edges(data=True)))
print("default graph attributes")
print(X.graph)
print("node node attributes")
print(X.nodes.data(True))

nx.draw(A)
plt.show()
```
